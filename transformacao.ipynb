```python
import pandas as pd
import pandera as pa
```

```python
valores_ausentes = ['**', '###!', '####', '****', '*****', 'NULL']
df = pd.read_csv("ocorrencia_2010_2020.csv", sep=";", parse_dates=['ocorrencia_dia'], dayfirst = True, na_values=valores_ausentes)
df.head(10)
```

```python
schema = pa.DataFrameSchema(
    columns = {
        "codigo_ocorrencia":pa.Column(pa.Int),
        "codigo_ocorrencia2":pa.Column(pa.Int),
        "ocorrencia_classificacao":pa.Column(pa.String),
        "ocorrencia_cidade":pa.Column(pa.String),
        "ocorrencia_uf":pa.Column(pa.String, pa.Check.str_length(2, 2), nullable=True),
        "ocorrencia_aerodromo":pa.Column(pa.String, nullable=True),
        "ocorrencia_dia":pa.Column(pa.DateTime),
        "ocorrencia_hora":pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True),
        "total_recomendacoes":pa.Column(pa.Int)
    }
)
```

```python
schema.validate(df)
```

combinando filtros tipo todas os acidentes graves no estado de SP usamos o operador logico e(&)

```python
filtroOcor = df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
filtroUf = df.ocorrencia_uf == 'SP'
df.loc[filtroOcor & filtroUf]
```

Ocorrencia cuja a classificação seja incidente OU incidente grave no estado de SP

```python
filtroOcor = (df.ocorrencia_classificacao == 'INCIDENTE GRAVE') | (df.ocorrencia_classificacao == 'INCIDENTE')
filtroUf = df.ocorrencia_uf == 'SP'
df.loc[filtroOcor & filtroUf]
```

fazendo a mesma operação usando o 'isin'

```python
filtroOcor = df.ocorrencia_classificacao.isin(['INCIDENTE GRAVE', 'INCIDENTE']) 
filtroUf = df.ocorrencia_uf == 'SP'
df.loc[filtroOcor & filtroUf]
```

unir coluna dia com a coluna hora:
1º problema: não se pode concatenar um dado do tipo data com um dado do tipo string entao devemos transformar a coluna data para string;
2º problema: ainda quero que o dados data esteja no formato data entao devenmos usar um parametro do pandas

```python
df['ocorrencia_dia_hora'] = pd.to_datetime(df.ocorrencia_dia.astype(str) + ' ' + df.ocorrencia_hora)
```

agrupamento!

criando um dataframe somente do ano de 2015 e mês 3

```python
filtroAno = df.ocorrencia_dia.dt.year == 2015
filtroMes = df.ocorrencia_dia.dt.month == 3
df2015m3 = df.loc[filtroAno & filtroMes]
df2015m3
```

vamos agrupar os dados para saber quantos acidentes foRAM nomeado como ACIDENTE, INCIDENTE E INCIDENTE GRAVE;

```python
df2015m3.groupby(['ocorrencia_classificacao']).codigo_ocorrencia.count()
```

ATENÇÃO!!! use esse metodo para sabendo que existe todos os dados, se estiver dados faltantes ele n vai agrupar corretamente. exemplo:

```python
df2015m3.groupby(['ocorrencia_classificacao']).ocorrencia_aerodromo.count()
```

veja q na classificação acidente foram contabilizado 5 sendo q foram 15 isso pq alguns acidentes não tinha informação do aerodromo entao foram excluidos da contagem; um metodo para contornar isso e usar o size

```python
df2015m3.groupby(['ocorrencia_classificacao']).size()
```
