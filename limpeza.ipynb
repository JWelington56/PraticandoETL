```python
import pandas as pd
```

```python
df = pd.read_csv("ocorrencia_2010_2020.csv", sep=";", parse_dates=['ocorrencia_dia'], dayfirst = True)
df.head()
```

#trabalhando com df.loc[]
df.loc[<linha>,<nome da coluna>]
vejamos alguns exemplos

```python
df.loc[1,'ocorrencia_cidade']
```

para exibir somente as linha 1 ha 3

```python
df.loc[1:3]
```

para retirar só linhas especificas, exemplo para observar as linhas 10 e a linha 40 somente;

```python
df.loc[[10,40]]
```

para observar uma coluna inteira somente

```python
df.loc[:,'ocorrencia_cidade']
```

verificando se a coluna tem valores unico pra linhas diferentes

```python
df.codigo_ocorrencia.is_unique
```

deixar uma coluna como indice

```python
df.set_index('codigo_ocorrencia', inplace=True)
```

```python
df.head()
```

```python
df.loc[40324]
```

voltando ao indice padrao

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

padronizando dados nulos

```python
df.loc[0,'ocorrencia_aerodromo'] = '****'
```

```python
df
```

exemplo de como alterar um dado: alterando a classificação se SP para GRAVE

```python
df.loc[df.ocorrencia_uf == 'SP', ['ocorrencia_classificacao']] = 'GRAVE'
```

```python
df
```

```python
df.loc[df.ocorrencia_aerodromo == '****', ['ocorrencia_aerodromo']] = pd.NA
```

```python
df.head()
```

Limpando de forma automatica

```python
df.replace(['**', '###!', '####', '****', '*****', 'NULL'], pd.NA, inplace=True)
```

calculando o somatório de dados faltantes

```python
df.isna().sum()
```

add dados as lacunas no exemplo colocando o '0'

```python
df.fillna(0, inplace=True)
```
